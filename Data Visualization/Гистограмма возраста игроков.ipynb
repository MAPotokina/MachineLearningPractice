```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('PlayerPersonalData.csv')
```

```python
#Гистограмма возраста игроков
#Гистограмма возраста укажет на доминирующий возраст и особенности распределения


plt.hist(df.Age, color='purple', bins=30, alpha=1)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Гистограмма возраста игроков')
plt.grid(True)
plt.xticks(range(16, 43, 2))
plt.axis([16, 43, 0, 1600])
```

```python
# Как можно видеть из полученной гистограммы, доминирующий возраст футболистов - 25 лет.
```
